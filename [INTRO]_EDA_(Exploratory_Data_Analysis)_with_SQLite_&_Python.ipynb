```python
# importing the library and viewing its version
import sqlite3 as sql3

print(sql3.version)
```

```python
# creating a database and initializing its connection
con = sql3.connect('my_db')
```

```python
# instantiating the object to connect to the database
cursor = con.cursor()
```

```python
# using the "cursor" object and the ".execute()" function to create the table and its respective fields
cursor.execute(
    '''
    CREATE TABLE IF NOT EXISTS personal_data(
        register_number INT NOT NULL UNIQUE PRIMARY KEY,
        first_name VARCHAR(30) NOT NULL,
        last_name VARCHAR(30) NOT NULL,
        date_birth DATE NOT NULL, 
        city VARCHAR(20),
        country VARCHAR(10) NOT NULL
    )
    '''
)

# commit
con.commit()
```

```python
# instantiating data in the table in the respective fields

## register_number, first_name, last_name, date_birth (m-d-Y), city & country
cursor.execute(
    '''
    INSERT INTO personal_data 
    VALUES(1, 'Tim', 'Berners-Lee', '06/08/1956', 'London', 'United Kingdom'
    )
    '''
)
```

```python
# performing a database query command
query = cursor.execute(
    '''
    SELECT * 
    FROM personal_data
    '''
).fetchall()
```

```python
# viewing the query result
display(query)
```

```python
# performing a query command to view the name of the table(s) present in the database
table_names = cursor.execute(
    '''
    SELECT name 
    FROM sqlite_master 
    WHERE type='table'
    '''
).fetchall()

# viewing the query result
table_names
```

```python
# applying a query filter on the table to return data from all fields in the table
data = cursor.execute(
    '''
    SELECT * 
    FROM personal_data
    '''
).fetchall()

# running a repeat loop to print all rows in the table 
for line in data:
    print(line)
```

```python
# viewing the name of the fields present in the table
colNames_Sheet1 = cursor.execute(
    '''
    SELECT * 
    FROM personal_data
    '''
)

for column in colNames_Sheet1.description:
    print(column[0])
```

```python
# counting the number of records in the table through the field "register_number"
count_ID = cursor.execute(
    '''
    SELECT
    COUNT(register_number)
    FROM personal_data
    '''
).fetchall()[0]

print(f'Total records: {count_ID[0]}')
```

```python
# counting the number of unique records in the table present in the "country" field 
# with the "DISTINCT" function
countries = cursor.execute(
    '''
    SELECT 
    DISTINCT country
    FROM personal_data
    '''
).fetchall()

countries
```

```python
# counting the number of unique records in the table present in the "country" field 
# with the "DISTINCT" function and grouping the data with the "GROUP BY" function
countries = cursor.execute(
    '''
    SELECT 
    DISTINCT *
    FROM personal_data
    GROUP BY country
    '''
).fetchall()

countries
```

```python
# viewing the first name in the "first_name" field and returning all data from the table
firstName = cursor.execute(
    '''
    SELECT
    DISTINCT first_name
    FROM personal_data
    '''
).fetchall()

firstName
```

```python
# applying a filter in the "country" field where it is equal to "United States"
countriesUnique = cursor.execute(
    '''
    SELECT *
    FROM personal_data
    WHERE country = 'United States'
    '''
).fetchall()

for line in countriesUnique:
    print(line)

# countriesUnique
```

```python
# performing the query on the table and instantiating the name of the fields in a list
colNames = cursor.execute(
    '''
    SELECT * 
    FROM personal_data
    '''
)

cols = []
for column in colNames.description:
    cols.append(column[0])

cols
```

```python
# viewing table data
data = cursor.execute(
    '''
    SELECT * 
    FROM personal_data
    '''
).fetchall()

for row in data:
    print(row)
```
